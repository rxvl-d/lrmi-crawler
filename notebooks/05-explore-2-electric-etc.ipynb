```python
import dask.distributed
import dask.dataframe
from dask.distributed import Client
```

```python
client = Client(n_workers=10, threads_per_worker=2, memory_limit="3GB")
client
```

```python
years = [2019, 2020, 2021]
```

```python
ddf = dict()
for year in years:
    ddf[year] = dask.dataframe.read_csv(f'../data/out/{year}/*.nq',
                                        names=['s', 'p', 'o', 'g'],
                                        blocksize='1024M')
```

```python
for year in years:
    ddf[year]['pld'] = ddf[year].s.str.extract('.*://(.+?)/.*')[0]
```
