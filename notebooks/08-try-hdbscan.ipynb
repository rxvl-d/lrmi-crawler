```python
import re

def extract_domain(url):
    pattern = r"(?<=://)([a-zA-Z0-9.-]+)"
    match = re.search(pattern, url)
    if match:
        return match.group(1)
    else:
        return None
```

```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
from matplotlib.colors import to_hex
from IPython.display import display

def style_dataframe(categorical_series, label_series, w=5, colormap='tab10'):
    assert len(categorical_series) == len(label_series), "Both series must have the same length."
    
    n = len(categorical_series)
    height = math.ceil(n / w)
    
    # Create an empty DataFrame with the most square shape possible
    df = pd.DataFrame(index=range(height), columns=range(w))
    
    # Fill the DataFrame with the categorical values
    for index, value in enumerate(categorical_series):
        row, col = divmod(index, w)
        df.at[row, col] = value
        
    # Assign labels to each value
    label_map = {}
    unique_labels = sorted(set(label_series))
    label_color_index = {label: index for index, label in enumerate(unique_labels)}

    for index, value in enumerate(categorical_series):
        label_map[value] = label_series[index]

    # Create a colormap with distinct colors
    cmap = sns.color_palette(colormap, n_colors=len(unique_labels))

    # Define the background color style function
    def background_color(val):
        if pd.isna(val):
            return 'background-color: #FFFFFF'
        else:
            color = to_hex(cmap[label_color_index[label_map[val]]])
            return f'background-color: {color}'

    # Apply the style function to the DataFrame and display it
    styled_df = df.style.applymap(background_color)
    display(styled_df)
    return df

# Sample usage
categorical_data = pd.Series(['apple', 'banana', 'apple', 'orange', 'banana', 'apple', 'grape', 'pear', 'melon'])
labels = pd.Series([5, 13, 5, 8, 13, 5, 8, 13, 8])

_ = style_dataframe(categorical_data, labels)
```

```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
from matplotlib.colors import to_hex
from IPython.display import display

def style_dataframe_grad(categorical_series, label_series_1, label_series_2, colormap='tab10'):
    assert len(categorical_series) == len(label_series_1) == len(label_series_2), "All series must have the same length."
    
    n = len(categorical_series)
    side_length = math.ceil(math.sqrt(n))
    
    # Create an empty DataFrame with the most square shape possible
    df = pd.DataFrame(index=range(side_length), columns=range(side_length))
    
    # Fill the DataFrame with the categorical values
    for index, value in enumerate(categorical_series):
        row, col = divmod(index, side_length)
        df.at[row, col] = value
        
    # Assign labels to each value
    label_map_1 = {}
    label_map_2 = {}
    unique_labels_1 = sorted(set(label_series_1))
    unique_labels_2 = sorted(set(label_series_2))
    label_color_index_1 = {label: index for index, label in enumerate(unique_labels_1)}
    label_color_index_2 = {label: index for index, label in enumerate(unique_labels_2)}

    for index, value in enumerate(categorical_series):
        label_map_1[value] = label_series_1[index]
        label_map_2[value] = label_series_2[index]

    # Create colormaps with distinct colors
    cmap_1 = sns.color_palette(colormap, n_colors=len(unique_labels_1))
    cmap_2 = sns.color_palette(colormap, n_colors=len(unique_labels_2), desat=0.5)

    # Define the background color style function
    def background_color(val):
        if pd.isna(val):
            return 'background-color: #FFFFFF'
        else:
            color1 = to_hex(cmap_1[label_color_index_1[label_map_1[val]]])
            color2 = to_hex(cmap_2[label_color_index_2[label_map_2[val]]])
            return f'background-image: linear-gradient(45deg, {color1} 50%, {color2} 50%)'

    # Apply the style function to the DataFrame and display it
    styled_df = df.style.applymap(background_color)
    display(styled_df)
    return df

# Sample usage
categorical_data = pd.Series(['apple', 'banana', 'apple', 'orange', 'banana', 'apple', 'grape', 'pear', 'melon'])
labels_1 = pd.Series([5, 13, 5, 8, 13, 5, 8, 13, 8])
labels_2 = pd.Series([1, 3, 1, 2, 3, 1, 2, 3, 2])

_ = style_dataframe_grad(categorical_data, labels_1, labels_2)
```
